```python
import requests
import json
import pandas as pd
from IPython.display import Image
```

```python
def getFromiTunes(path,queryParams=dict()):
    url = f"https://itunes.apple.com{path}"
    res = requests.get(url, params=queryParams)
    print(res.status_code, res.url)
    return res.json()
```

```python
queryParams = {
        "term":"Elvis Presley",
        "entity":"song",
        "limit": 100
    }
iTunes = getFromiTunes("/search",queryParams)["results"]
iTunes
```
